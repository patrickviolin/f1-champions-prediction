## Initial Exploratory Data Analysis

```python
import numpy as np
import pandas as pd

results = pd.read_csv("../data/01_raw/results.csv")
races = pd.read_csv("../data/01_raw/races.csv")
drivers = pd.read_csv("../data/01_raw/drivers.csv")
constructors = pd.read_csv("../data/01_raw/constructors.csv")
status = pd.read_csv("../data/01_raw/status.csv")
```

```python
# Verifying outliers
results_status = pd.merge(left=status, right=results, left_on="statusId", right_on="statusId", how="inner",
                          validate="one_to_many")
results_status[["statusId", "status", "number"]].groupby(by=["statusId", "status"]).count()
```

```python
# Verifying statuses who disqualify
results_status['is_dnf'] = np.where(results_status['position'] == '\\N', 'DNF (\\N)', 'Finished')

is_dnf_count = results_status.groupby(['statusId', 'status', 'is_dnf']).size().reset_index(name='total_occurrences')

view_dnf_count_table = is_dnf_count.pivot_table(
    index=['statusId', 'status'],
    columns='is_dnf',
    values='total_occurrences',
    fill_value=0
)

view_dnf_count_table
```

 ## Joining Data

```python
results_races = pd.merge(left=races, right=results, left_on="raceId", right_on="raceId", how="inner",
                         validate="one_to_many")
results_races_drivers = pd.merge(left=drivers, right=results_races, left_on="driverId", right_on="driverId",
                                 how="inner", validate="one_to_many")
results_races_drivers_constructors = pd.merge(left=constructors, right=results_races_drivers, left_on="constructorId",
                                              right_on="constructorId", how="inner", validate="one_to_many")
data_post_merge = pd.merge(left=status, right=results_races_drivers_constructors, left_on="statusId",
                           right_on="statusId", how="inner", validate="one_to_many")

data_post_merge.to_csv("../data/02_interim/results_races_drivers_constructors.csv", index=False)
```
